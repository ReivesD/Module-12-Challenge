### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}
result = establishments.find(query)

matching_establishments = list(result)

# Use count_documents to display the number of documents in the result
document_count = establishments.count_documents(query)

print("Number of matching documents:", document_count)

# Display the first document in the results using pprint
if matching_establishments:
    pprint(matching_establishments[0])
else:
    print("No matching documents found.")
```

```python
# Convert the result to a Pandas DataFrame
hygiene20_df = pd.DataFrame(matching_establishments)

# Display the number of rows in the DataFrame
num_rows = hygiene20_df.shape[0]

# Display the first 10 rows of the DataFrame
hygiene20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and RatingValue greater than or equal to "4"
london_authorities = [
    "Barking and Dagenham", "Bexley", "Brentwood", "Bromley", "City of London Corporation",
    "Dartford", "Greenwich", "Hackney", "Havering", "Kensington and Chelsea", "Lambeth",
    "Lewisham", "Maldon", "Newham", "Redbridge", "Rochford", "Salford", "Slough",
    "Tower Hamlets", "Waltham Forest"
]

query = {
    "LocalAuthorityName": {"$in": london_authorities},
    "RatingValue": {"$gte": "4"}
}
result = establishments.find(query)

# Use count_documents to display the number of documents in the result
document_count = establishments.count_documents(query)
print("Number of matching documents:", document_count)

# Display the first document in the results using pprint
if document_count > 0:
    first_document = next(result)
    pprint(first_document)
else:
    print("No matching documents found.")
```

```python
# Convert the result to a Pandas DataFrame
ratings4_df = pd.DataFrame(list(result))

# Display the number of rows in the DataFrame
num_rows = ratings4_df.shape[0]
print("Number of rows in the DataFrame:", num_rows)

# Display the first 10 rows of the DataFrame
ratings4_df.head(10)
```
